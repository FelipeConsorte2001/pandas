```python
import pandas as pd
from google.colab import drive
import os
drive.mount('/content/gdrive', force_remount=True)
path = 'gdrive/My Drive/pandas/vendas.xlsx'

vendas_df = pd.read_excel(f'{path}')
display(vendas_df.describe())
```

```python
produto = vendas_df[['Produto', 'Quantidade']]
display(produto)
```

```python
# PEgar uma/varias linhas
display(vendas_df.loc[1:10])

# Pega linhas que correspondem uma unica maneira
vendas_shopping = display(vendas_df.loc[vendas_df['ID Loja'] == 'Shopping Morumbi'])
display(vendas_shopping)

# Pegar varias linhas e colunas usando o loc
vendas_shopping = display(vendas_df.loc[vendas_df['ID Loja'] == 'Shopping Morumbi'], ['ID Loja', 'Produto', 'Quantidade'])
display(vendas_shopping)


# Pega linha e valor especifico
print(vendas_df.loc[1, 'Produto'])
```

```python
# Adicionar coluna a partir de uma existente
vendas_df['Comissão'] = vendas_df['Valor Final'] * 0.05
display(vendas_df)

# Cria coluna a com valor padrão
vendas_df.loc[:, 'Imposto'] = 0
display(vendas_df)
```

```python
path_dez = 'gdrive/My Drive/pandas/vendas-dez.xlsx'
vendas_dez_df = pd.read_excel(f'{path_dez}')
vendas_df = vendas_df.append(vendas_dez_df) # adicionar valor uma planilia pra outra 
vendas_df.loc[:, 'Imposto'] = 0
vendas_df.loc[:, ['Comissão']] = vendas_df['Valor Final'] * 0.05
display(vendas_df)
```

```python
# Excluir coluna e linha
vendas_df = vendas_df.drop('Imposto', axis=1) # 1 para coluna e 0 para linha 
display(vendas_df)
```

```python
# Deletar linhas e colunas que estão vazias
vendas_df = vendas_df.dropna(how='all', axis=)

# excluir linhas que possuem pelo menos  valor vazio 
vendas_df = vendas_df.dropna()

# preencher valores vazios
# preencher com a media da coluna
vendas_df['Comissão'] = vendas_df['Comissão'].fillna(vendas_df['Comissão'].mean())
display(vendas_df)

# preencher ultimo valor
vendas_df = vendas_df.ffill()
```

```python
# Value conts = soma os valores igual em uma coluna
trasnsacoes_loja = vendas_df['ID Loja'].value_counts()
#display(trasnsacoes_loja)

# gruop by
faturamento_produto = vendas_df[['Produto', 'Valor Final']].groupby('Produto').mean()
display(faturamento_produto)
```

```python
# Mesclar tabela
gerentes_df = pd.read_excel('gdrive/My Drive/pandas/gerentes.xlsx')

vendas_df = vendas_df.merge(gerentes_df)
display(vendas_df)
```
